# U.S. Medical Insurance Costs

```python
#Import csv module
import csv

#Import data to python list as list of lists
medical_data = []
with open("insurance.csv") as raw_data_csv:
    data = csv.DictReader(raw_data_csv, delimiter = ",")
    for item in data:
        medical_data.append(item)
#print(medical_data)

#Modify smoker and gender to quantitative factors

def cleanup(medical_data):
    for person in medical_data:
        if person["sex"] == "male":
            person["sex"] = 1
        else:
            person["sex"] = 0

        if person["smoker"] == "yes":
            person["smoker"] = 1
        else:
            person["smoker"] = 0
        
        person["bmi"] = float(person["bmi"])
        person["age"] = int(person["age"])
        person["children"] = int(person["children"])
        person["charges"] = float(person["charges"])
    return medical_data

updated_medical_date = cleanup(medical_data)
#print(updated_medical_date)
```

```python
#Extract variable as a list
def variable_extractor(data, variable): #Takes a list of medical records as dictionaries, string variable name
    var = []
    for details in data:
        var.append(details[variable])
    return var

#Calculate the average age
ages = variable_extractor(updated_medical_date, "age")

def average_age_calc(ages): #Function takes a list of ages
    total_ages = 0
    for age in ages:
        total_ages += age
    return total_ages / len(ages)
#test
#print(average_age_calc(ages))

#Where most persons come from
regions = variable_extractor(updated_medical_date, "region")

def area_with_most_ppl_calc(regions): #arguments: list of locations
    loc = ""
    locs = {}
    for region in regions:
        if region in locs.keys():
            locs[region] += 1
        else:
            locs[region] = 1
    for region in locs.keys():
        if locs[region] == max(locs.values()):
            loc = region
        
    return loc  
#test
#print(area_with_most_ppl_calc(regions))

#Average age of people with atleast x kid(s)
def average_age_with_kids(data, x): #Takes list of dictionaries and number of kids as int
    try:
        total_age = 0
        count = 0
        for person in data:
            if int(person["children"]) >= x:
                total_age += int(person["age"])
                count += 1
        return total_age / count
    except:
        return 0
#test
#print(average_age_with_kids(medical_data, 1))
```

```python
#get correlation coefficient for quantifiable variables
from scipy.stats import pearsonr

def correlation_calculator(data, variable): #takes list of dictionaries and a string variable
    y = variable_extractor(updated_medical_date, "charges")
    x = variable_extractor(updated_medical_date, variable)

    corr,p = pearsonr(x, y)
    
    if corr < -0.75:
        relation = "strong negative correlation"
    elif corr < - 0.25:
        relation = "negative correlation"
    elif corr < 0:
        relation = "weak negative correlation"
    elif corr == 0:
        relation = "no correlation"
    elif corr < 0.25:
        relation = "weak positive correlation"
    elif corr < 0.75:
        relation = "positive correlation"
    else:
        relation = "strong positive correlation"
       
    return variable + " has a " + relation + " with a charges with a correlation coefficient of : " + str(corr)

#Test
print(correlation_calculator(updated_medical_date, "age"))
print(correlation_calculator(updated_medical_date, "bmi"))
print(correlation_calculator(updated_medical_date, "smoker"))
print(correlation_calculator(updated_medical_date, "children"))
print(correlation_calculator(updated_medical_date, "sex"))
        
```
